```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
# getting the statistical measure of the data
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

0 --> non diabetes
1 --> diabetes

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
#separating the data and labels
X = diabetes_dataset.drop(columns = 'Outcome',axis = 1)    # axis = 1 means drooping column & axis = 0 means droping row
Y= diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Data standardization
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

```python
# spliting data into training data and test data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training the model

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
# Training the support vector machine classifier
classifier.fit(X_train,Y_train)
```

Model Evalution

Accuracy score

```python
# accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print("accuracy score of training data : ", training_data_accuracy )
```

```python
# Accuracy score of test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of test data : ',test_data_accuracy)
```

Making a predictive system

```python
input_data = (63,14,7,44,15,58,0.62,51)

# changing the input data to a numpy array

input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance

input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

# standadize the input data

std_data = scaler.transform(input_data_reshape)
print(std_data)
```

```python
prediction = classifier.predict(std_data)
```
